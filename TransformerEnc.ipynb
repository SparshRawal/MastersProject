```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.feature_extraction.text import TfidfVectorizer



# Example input data
input_sequences = []
for seq in X_train:
    input_sequences.append(np.array(seq.split(' ')))
input_sequences = np.array(input_sequences)

padded_seq = []
max_len = 0
for seq in input_sequences:
    if len(seq)>max_len:
        max_len=len(seq)
print("Max Length : ",max_len)
vectorizer = TfidfVectorizer(ngram_range=(1,1),max_features=30)

X_train_tfidf = vectorizer.fit_transform(X_train)

X_train_tfidf = X_train_tfidf.toarray()
print(X_train_tfidf.shape)

target_languages = y_train_ind # Example target languages (0: English, 1: Other language)

target_languages = torch.tensor(target_languages)
input_sequences = nn.utils.rnn.pad_sequence([torch.from_numpy(seq).float() for seq in X_train_onehot], batch_first=True)
validation_input = nn.utils.rnn.pad_sequence([torch.from_numpy(seq).float() for seq in X_dev_onehot], batch_first=True)

print(input_sequences.shape)
print(validation_input.shape)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

# Custom Dataset class
class MyDataset(Dataset):
    def __init__(self, data, targets):
        self.data = data
        self.targets = targets

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = self.data[idx]
        y = self.targets[idx]
        return x, y

# Example usage
data = input_sequences  # Your input data as a NumPy array or Python list
targets = y_train_one_hot  # Your target labels as a NumPy array or Python list

# Convert data and targets to tensors
data_tensor = torch.Tensor(data).cuda()
targets_tensor = torch.Tensor(targets).cuda()
Yt_train = targets_tensor.type(torch.LongTensor).cuda()


# Create an instance of the custom dataset
dataset = MyDataset(data_tensor, Yt_train)

# Set the batch size and other DataLoader parameters
batch_size = 50
shuffle = False

# Create the data loader
train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)

input = validation_input
target = y_dev_one_hot

input_tensor = torch.Tensor(input).cuda()
output_tensor = torch.Tensor(target).cuda()
Yt_Dev = output_tensor.type(torch.LongTensor).cuda()

val_dataset = MyDataset(input_tensor,Yt_Dev)
val_loader = DataLoader(val_dataset,batch_size=batch_size, shuffle=shuffle)


```

```python

# # Define the Transformer Encoder-Decoder Model
class TransformerEncoderDecoder(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_dim, num_layers, num_heads, dropout):
        super(TransformerEncoderDecoder, self).__init__()

        # Encoder layers
        self.encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(
                d_model=input_dim,
                nhead=num_heads,
                dim_feedforward=hidden_dim,
                dropout=dropout
            ),
            num_layers
        )

        # Decoder layers
        self.decoder = nn.TransformerDecoder(
            nn.TransformerDecoderLayer(
                d_model=input_dim,
                nhead=num_heads,
                dim_feedforward=hidden_dim,
                dropout=dropout
            ),
            num_layers
        )

        # Output linear layer
        self.linear = nn.Linear(input_dim, output_dim)

    def forward(self, src):
        encoder_output = self.encoder(src)
        decoder_output = self.decoder(encoder_output, encoder_output)
        output = self.linear(decoder_output)
        return output


# Define the hyperparameters
input_dim = 128 # Dimensionality of the input sequences after one-hot encoding
output_dim = 20  # Number of target languages
hidden_dim = 512  # Hidden dimension of the model
num_layers = 8  # Number of layers in the encoder and decoder
num_heads = 16  # Number of attention heads
dropout = 0.1  # Dropout rate

# Create an instance of the Transformer Encoder-Decoder model
modelTransformer = TransformerEncoderDecoder(input_dim, output_dim, hidden_dim, num_layers, num_heads, dropout).cuda()

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(modelTransformer.parameters(), lr=0.01)
```

```python
def train_miniBatch(modelTransformer, optimizer, criterion, num_epochs):
    # Set the number of mini-batches to accumulate gradients
    patience = 5
    # Training loop
    best_loss = float('inf')
    for epoch in range(num_epochs):
        total_loss = 0
        total_correct = 0
        total_samples = 0
        modelTransformer.train()
        
        for batch_idx, (data, targets) in enumerate(train_loader):
            # Forward pass
            output = modelTransformer(data)
            loss = criterion(output, targets)

            # Backward pass
            output = torch.max(output, 1)
            preds = output.values.argmax(1)
            targs = targets.argmax(1)
   
            new  = (preds == targs).sum().item()

            total_correct += new

            loss.backward()


            total_samples += targets.size(0)
            print("Batch : " ,batch_idx," Total Correct : ",total_correct," Total Samples : ",total_samples," Accuracy : ",total_correct/total_samples,end='\r')

            optimizer.step()

        modelTransformer.eval()
        val_loss = 0
        with torch.no_grad():
            
            for val_data, val_targets in val_loader:
                val_output = modelTransformer(val_data)
                val_batch_loss = criterion(val_output, val_targets)
                val_loss += val_batch_loss.item() * val_data.size(0)
            val_loss /= len(val_loader.dataset)

    # Check stopping criteria
        if val_loss < best_loss:
            best_loss = val_loss
            counter = 0
        else:
            counter += 1
            if counter >= patience:
                print("Early stopping. No improvement in validation loss.")
                break
        # Compute the average loss and accuracy for the epoch
        
        average_loss = total_loss / total_samples
        accuracy = total_correct / total_samples

        # Print or log the average loss and accuracy for the epoch
        print()
        print()
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {average_loss:.4f}, Accuracy: {accuracy:.4f}, Validation Loss: {val_loss:.4f}")

# Example usage
num_epochs = 10
```

```python
train_miniBatch(modelTransformer, optimizer, criterion, 10)
```
